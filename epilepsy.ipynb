```python

import glob

nii_files = glob.glob("*.nii") #create a list with file names with .nii extension
anatomical = nii_files[0] #variable containing anatomical/background file name
functional = nii_files[1] #varianble containing functional file name
```

```python
from nilearn import plotting
#plot function from nilearn.plotting with custom color map, title and coordinates
plotting.plot_stat_map(functional, anatomical, title='Epilepsy', cmap='ocean', threshold=0.1, cut_coords=(12, 18, 20))

```

```python
import nibabel as nib
import matplotlib.pyplot as plt
img = nib.load(functional) #load the functional volume using the nib.load function
```

```python

data = img.get_fdata() #extracting the data as an array 
positive_values = data[data > 0] #selecting only positive values 

# Plot the histogram using matplotlib
plt.figure(figsize=(20, 5))
plt.hist(positive_values, bins=12, color='black', alpha=0.7)
plt.title("fMRI signal distribution")
plt.xlabel("intensity")
plt.ylabel("frequency")
plt.show()
```
